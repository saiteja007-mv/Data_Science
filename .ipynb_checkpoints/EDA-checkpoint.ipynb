# Step 2: Data Preparation
* Dropping Irrelevant Columns and Rows
* Identify duplicated columns
* Renaming Columns
* Feature Columns

```python
# We can also use drop function for this dropping but subsetting and selecting rows would be better
df = df[['coaster_name', 
    #'Length', 'Speed', 
    'Location', 'Status',
    #'Opening date',
     #  'Type', 
    'Manufacturer', 
         #'Height restriction', 'Model', 'Height',
      # 'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
     #  'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
     #  'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
     #  'Track layout', 'Fastrack available', 'Soft opening date.1',
     #  'Closing date', 
    #'Opened', 
    # 'Replaced by', 'Website',
    #   'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
    #   'Single rider line available', 'Restraint Style',
     #  'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 
    'latitude', 'longitude', 
    'Type_Main',
       'opening_date_clean',
    #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
    #'height_value', 'height_unit', 
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['opening_date_clean']=pd.to_datetime(df['opening_date_clean'])
```

```python
#Renaming Columns
df = df.rename(columns={'coaster_name':'Coaster_Name',
                   'year_introduced': 'Year_Introduced',
                  'opening_date_clean':'Opening_Date',
                  'speed_mph': 'Speed_mph',
                  'Inversions_clean':'Inversions',
                   'height_ft':'Height_ft',
                  'Gforce_clean':'GForce'})
```

```python
# Missing values
df.isna().sum()
```

```python
df.dropna()
```

```python
# Duplicate data
df.duplicated().sum()
```

```python
df.columns
```

```python
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Manufacturer', 'Opening_Date','latitude', 'longitude', 'Speed_mph','Height_ft', 'GForce', 'Status'])].reset_index(drop=True)
```

```python
df
```

```python
df.shape
```

```python
df.columns
```

# Step 3: Feature Understanding
#### (Univariate Analysis)
* Plotting Feature Distributions
    * Histograms
    * KDE
    * Boxplot


```python
 df['Year_Introduced'].value_counts() # How many Coaster are introduced in Years
```

```python
ax = df['Year_Introduced'].value_counts().head(10).plot(kind='bar',title='Coasters Introduced Over Years')
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')
plt.savefig('Graphs/EDA/Coasters_Introduced_OverYear.png',dpi=300)

# plt.figure(figsize=(8,5))

# plt.(df.Year_Introduced.unique(),df.Year_Introduced.value_counts())
```

```python
ax = df['Speed_mph'].plot(kind='hist',bins=20,title='Coaster Speed (mph)')
ax.set_xlabel('Speed (mph)')
plt.savefig('Graphs/EDA/Coaster_Speed_Frequency.png',dpi=300)
```

```python
ax = df['Speed_mph'].plot(kind='kde',title='Coaster Speed (mph)')
ax.set_xlabel('Speed (mph)')
plt.savefig('Graphs/EDA/Coaster_Speed_Frequency_Distribution.png',dpi=300)
```

# Step 4: Feature Relationships
* Scatterplot
* Heatmap Correlation
* Pairplot
* Groupby Comparisons

```python
df.plot(kind='scatter',x='Speed_mph',y='Height_ft',title='Coaster Speed vs Height')
plt.savefig('Graphs/EDA/Scatter_Coasterspeed_height.png',dpi=300)
plt.show()
```

```python
sns.scatterplot(x='Speed_mph',y='Height_ft',data=df,hue='Year_Introduced')
plt.savefig('Graphs/EDA/Scatter_Coasterspeed_height_HUE.png',dpi=300)
plt.show()
```

```python
df.columns
```

```python
sns.pairplot(df,vars=['Year_Introduced', 'Speed_mph',
       'Height_ft', 'Inversions', 'GForce'],hue='Type_Main')
plt.savefig('Graphs/EDA/Pairplot.png',dpi=300)
plt.show()
```

```python
df_corr = df[['Year_Introduced', 'Speed_mph',
       'Height_ft', 'Inversions', 'GForce']].dropna().corr()
```

```python
sns.heatmap(df_corr,annot=True)
plt.savefig('Graphs/EDA/Heatmap_correlation.png',dpi=300)
```
