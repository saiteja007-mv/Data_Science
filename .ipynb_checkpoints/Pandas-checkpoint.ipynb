### Making changes to the data

#### Adding a column into the data

```python
df['Total'] = df['HP']+df['Attack']+df['Defense']+df['Sp. Atk']+df['Sp. Def']+df['Speed'] #We can directly add the column
df.head(5)
```

#### Droping the column

```python
df = df.drop(columns = ['Total'])
df.head(2)
```

```python
df['Total'] = df.iloc[:,4:10].sum(axis=1)  #axis =1 means it will add horizontally
df.head(3)
```

```python
#We we have to move the added column to some other column place
cols = list(df.columns)
df = df[cols[0:4]+[cols[-1]]+cols[4:12]]
df.head(2)
```

# Filtering Data

```python
df_hp = df.loc[(df['HP']>120) & (df['Type 1']=='Water')] #Here and is % or is |
```

```python
df_hp.head(3)
```

```python
#If the obtain data index is clumsy we can use reset_index 
df_hp.reset_index(drop = True,inplace = True)
df_hp
```

```python
#Some examples 
#If you want to get the data that contain particular string
df.loc[df['Name'].str.contains('Mega')].head(3)
```

```python
#Using regular expression
import re
df.loc[df['Type 1'].str.contains('n[a-z]*',flags = re.I,regex = True)] #flags can give output whether the letter is capital or not
```

# Conditional Changes

```python
df.loc[df['Total']> 500, ['Generation','Legendary']] = ['Test 1','Test 2']  #df.loc[condition, columns] = [Update values]
```

```python
df
```

# Aggregate Statistics ( Group by )

```python
df.groupby('Type 1')[['HP','Attack']].mean().sort_values('HP')  # groupby(columns_to_be_grouped)[columns_for_aggregate]
```

```python
df.groupby('Type 1')[['HP','Attack']].sum()
```

```python
df.groupby(['Type 1','Type 2'])[['HP','Attack']].count() 
```

# Working with large amount of data

```python
count = 0
for df in pd.read_csv('data/pokemon_data.csv',chunksize = 5):
    count += 1
print(count)
```

```python
new_df = pd.DataFrame(columns = df.columns)
for df in pd.read_csv('data/pokemon_data.csv',chunksize = 5):
    results = df.groupby('Type 1').count()
    new_df = pd.concat([new_df,results])
new_df.head(5)
new_df.shape
```
